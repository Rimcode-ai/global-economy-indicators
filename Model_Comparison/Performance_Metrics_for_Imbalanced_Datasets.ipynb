```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the metrics for the models
data = {
    "Model": [
        "Logistic Regression", "Decision Tree", "Random Forest",
        "SVM", "AdaBoost", "Gradient Boosting", "XGBoost", "Naive Bayes"
    ],
    "Precision (1)": [0.751429, 0.780269, 0.768177, 0.751071, 0.754114, 0.757649, 0.772775, 0.745902],
    "Recall (1)": [0.999367, 0.771374, 0.923369, 0.999367, 0.986700, 0.987967, 0.934769, 0.057631],
    "F1-Score (1)": [0.857842, 0.775796, 0.838654, 0.857609, 0.854870, 0.857614, 0.846088, 0.106996]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the model name as the index
df.set_index("Model", inplace=True)

# Plot Precision, Recall, and F1-Score for each model
#plt.figure(figsize=(12, 6))
df[["Precision (1)", "Recall (1)", "F1-Score (1)"]].plot.bar(rot=45, figsize=(12, 6))
plt.title("Performance Metrics for Imbalanced Datasets")
plt.ylabel("Score")
plt.xlabel("Model")
plt.legend(title="Metrics", loc="lower right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
