```python
import pandas as pd
import matplotlib.pyplot as plt

# Classification model data
classification_data = {
    "Model": [
        "Logistic Regression", "Decision Tree", "Random Forest",
        "SVM", "AdaBoost", "Gradient Boosting", "XGBoost", "Naive Bayes"
    ],
    "Accuracy": [0.751308, 0.665240, 0.733238, 0.750832, 0.748455, 0.753685, 0.744650, 0.277699],
    "Precision (1)": [0.751429, 0.780269, 0.768177, 0.751071, 0.754114, 0.757649, 0.772775, 0.745902],
    "Recall (1)": [0.999367, 0.771374, 0.923369, 0.999367, 0.986700, 0.987967, 0.934769, 0.057631],
    "F1-Score (1)": [0.857842, 0.775796, 0.838654, 0.857609, 0.854870, 0.857614, 0.846088, 0.106996]
}

classification_df = pd.DataFrame(classification_data)

# Regression model data
regression_data = {
    "Model": ["Linear Regression", "Ridge Regression", "Lasso Regression"],
    "MSE": [2.545620e+21, 2.505705e+21, 2.531838e+21],
    "MAE": [1.233265e+10, 1.222778e+10, 1.208694e+10],
    "R-squared": [0.133737, 0.147320, 0.138427]
}

regression_df = pd.DataFrame(regression_data)

# Visualization for classification metrics
#plt.figure(figsize=(12, 6))
classification_df.set_index("Model")[["Accuracy", "F1-Score (1)"]].plot.bar(rot=45)
plt.title("Classification Model Performance")
plt.ylabel("Score")
plt.xlabel("Model")
plt.legend(loc="lower right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Visualization for regression metrics
#plt.figure(figsize=(12, 6))
regression_df.set_index("Model")[["MSE", "MAE"]].plot.bar(rot=45)
plt.title("Regression Model Error Metrics")
plt.ylabel("Error")
plt.xlabel("Model")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

plt.figure(figsize=(8, 6))
regression_df.set_index("Model")["R-squared"].plot.bar(rot=45, color='orange')
plt.title("Regression Model R-squared")
plt.ylabel("R-squared")
plt.xlabel("Model")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
